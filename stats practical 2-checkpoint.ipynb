# In probability theory, a log-normal (or lognormal) distribution is a continuous probability distribution of a random variable whose logarithm is normally distributed. Thus, if the random variable X is log-normally distributed, then Y = ln(X) has a normal distribution.[2][3] Equivalently, if Y has a normal distribution, then the exponential function of Y, X = exp(Y), has a log-normal distribution.

```python
mu= 3.0
sigma=1.0

lognormal_distributed_data=np.random.lognormal(mu,sigma,1000)
```

```python

sns.histplot(lognormal_distributed_data,kde=True)

# This is lognomral distribution
```

```python
sns.histplot(np.log(lognormal_distributed_data),kde=True)

# np.log command  convert the log normal distributed data to normal distributed data

# if you have outliers in your data than your to take log to convert  from the log normal distribution of the data to 
# normal distribution data
```

# QQ plot

```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab

def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample,kde=True)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist="norm",plot=pylab)
    plt.show()
```

```python
plot_data(lognormal_distributed_data)
```

```python
plot_data(np.log(lognormal_distributed_data))
```

suppose you have population with any shape of probability distribution (not necessarily normal),and you draw random samples of a
fixed size from this population.As the sample size increases,the distribution of the sample means approches a normal distribution,
regardless of the shape of the original population distribution.

```python
sample1=np.random.choice(population,20)
sample2=np.random.choice(population,20)
sample3=np.random.choice(population,20)
sample4=np.random.choice(population,20)
```

```python
print(sample1)
print(60*"=")
print(sample2)
print(60*"=")
print(sample3)
print(60*"=")
print(sample4)
print(60*"=")
```

```python
mean_of_the_sample=[]

all_sample=[sample1,sample2,sample3,sample4]

for sample in all_sample:
    print(np.mean(sample))
    mean_of_the_sample.append(np.mean(sample))
```
